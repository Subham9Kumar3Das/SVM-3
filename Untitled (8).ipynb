* 1. Wx+b is a linear SVM function.
* 2. ||W||/2 is the obective function of linear SVM.
* 3. Kernel is function used to convert the graph into higher dimensional graph so the dataset can be easily separable by a hyperplane.
* 4. Support vectors in SVM are like critical examples that are crucial for defining the decision boundary. Imagine you have a bunch of points on a football field, and you want to draw a line to separate players on two teams. The support vectors are the players standing right at the edge of the field; they're critical for deciding where to draw the line (decision boundary). Remove them, and your line might shift, affecting the whole game plan. In SVM, these support vectors determine the position of the decision boundary and are vital for accurate classification.
* 5. Hyperplane: Think of it as the magic line that separates things. In a 2D world, it's a line; in 3D, it's a plane.

     Marginal Plane: Imagine two parallel sidelines next to the magic line. These sidelines help define a safe zone around the line.

     Soft Margin: It's like allowing a few mistakes. Some points can be on the wrong side, but not too many.

     Hard Margin: No mistakes allowed. All points must be on the right side of the magic line, no exceptions.

```python
#6.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
x=pd.DataFrame(iris.data,columns=iris.feature_names)
y=pd.DataFrame(iris.target,columns=['Outcome'])
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
svc=SVC(kernel='linear')
svc
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
svc.fit(x_train,y_train)
```

```python
y_pred=svc.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
svc.coef_,svc.intercept_
```

# Hyperparameter Tunning.

```python
# Hyperparameter Tunning.
paramater={'C':[0.1,1,10,100,1000],
          'gamma':[1,0.1,0.01,0.001,0.0001],
           'kernel':['linear','poly','rbf','sigmoid']
          }
```

```python
grid=GridSearchCV(SVC(),param_grid=paramater,cv=5,refit=True,verbose=3)
grid.fit(x_train,y_train)
```
